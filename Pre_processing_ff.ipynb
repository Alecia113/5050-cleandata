```python
import torch
from nltk.tokenize import TweetTokenizer, word_tokenize
import nltk
```

```python
Token = TweetTokenizer()
for i in range(len(Delete_Displace)):
  try:
    Delete_Displace[i][-3] = Token.tokenize(Delete_Displace[i][-3])  
  except:
    continue
Delete_Displace[:2]
```

```python
Delete_Displace[15][-3]
```

```python
Smy = ';'
for i in range(len(Delete_Displace)):
  try:
    for j in range(len(Delete_Displace[i][-3])):
      if Delete_Displace[i][-3][j] == 'and':
        Delete_Displace[i][-3][j] = Smy
      elif Delete_Displace[i][-3][j] == '/':
        Delete_Displace[i][-3][j] = Smy
      elif Delete_Displace[i][-3][j] == 'or':
        Delete_Displace[i][-3][j] = Smy
      elif Delete_Displace[i][-3][j] == ',':
        Delete_Displace[i][-3][j] = Smy
      else:
        continue
  except:
    continue
Delete_Displace[:5]
```

```python
Block = ' '
Del_Disp = []
for i in range(len(Delete_Displace)):
    try:
        number = 0
        if number <= len(Delete_Displace[i][-3]):
            Delp = ''
            for j in range(len(Delete_Displace[i][-3])):
                Delp += Delete_Displace[i][-3][j]+ Block
            #print(Delp)
            Delete_Displace[i][-3] = Delp
        else:
            continue
    except:
        continue
        
```
